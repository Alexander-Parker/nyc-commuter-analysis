```python
import pandas as pd
```

```python
import pickle
```

```python
census = pd.read_csv('data/NY_census_data.csv', skiprows=1)
```

```python
census.info()
```

```python
census.head()
```

```python
census_dem = census[['Geography','Percent; SEX AND AGE - Female population','Percent; RACE - Total population - One Race - White']]
```

```python
census_dem.head()
```

```python
census_dem = census_dem.rename(columns={'Percent; SEX AND AGE - Female population':'Percent_Female'})
```

```python
census_dem = census_dem.rename(columns={'Percent; RACE - Total population - One Race - White':'Percent_White'})
```

```python
census_dem.head(20)
```

```python
census_dem = census_dem.rename(columns={'Geography':'Zip_Code'})
```

```python
census_dem['Percent_Female'] = pd.to_numeric(census_dem['Percent_Female'], errors='coerce')
```

```python
census_dem['Percent_White'] = pd.to_numeric(census_dem['Percent_White'], errors='coerce')
```

```python
census_dem['Percent_Non_White'] = 1 - census_dem['Percent_White']
```

```python
census_dem.head()
```

```python
census_dem[['DROP', 'ZIP']] = census_dem['Zip_Code'].str.split(expand=True)
```

```python
census_dem.head()
```

```python
census_dem.dtypes
```

```python
census_dem['Percent_Female'] = census_dem['Percent_Female'].div(100)
```

```python
census_dem['Percent_White'] = census_dem['Percent_White'].div(100)
```

```python
census_dem.head()
```

```python
census_clean = census_dem[['ZIP', 'Percent_Female', 'Percent_White','Percent_Non_White']]
```

```python
census_clean.head()
```

```python
with open('pickle/census_clean.pickle', 'wb') as to_write:
    pickle.dump(census_clean, to_write)
```
