```python
import pandas as pd
import pickle
import seaborn as sns
```

```python
with open('pickle/mta_data_intra_w.pickle', 'rb') as to_read: #run weighted_delta to generate
    mta_intra_weight = pickle.load(to_read)
    
with open('pickle/mta_data_daily_w.pickle', 'rb') as to_read: #run weighted_delta to generate
    mta_daily_weight = pickle.load(to_read)
```

```python
mta_daily_weight.head()
```

```python
mta_daily_weight.sort_values('Weighted_Traffic', ascending = False).head(20)
```

```python
wgt_daily_station = mta_daily_weight.groupby(['STATION', 'DATE'])['Weighted_Traffic'].sum().reset_index()
```

```python
wgt_daily_avg = wgt_daily_station.groupby(['STATION'])['Weighted_Traffic'].mean().reset_index()
```

```python
wgt_daily_top = wgt_daily_avg.sort_values('Weighted_Traffic', ascending=False).reset_index(drop=True)
wgt_daily_top.head()
```

```python
with open('pickle/wgt_daily_top.pickle', 'wb') as to_write:
    pickle.dump(wgt_daily_top, to_write)
```
